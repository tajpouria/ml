```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use("./deeplearning.mplstyle")
```

```python
x_train = np.array(
    [
        1.3,
        2.4,
        3.5,
        4.6,
        5.7,
        6.8,
        7.9,
        9.0,
        10.1,
        11.2,
        12.3,
        13.4,
        14.5,
        15.6,
        16.7,
        17.8,
        18.9,
        20.0,
    ]
)
y_train = np.array(
    [
        350.0,
        400.0,
        450.0,
        500.0,
        550.0,
        600.0,
        650.0,
        700.0,
        750.0,
        800.0,
        850.0,
        900.0,
        950.0,
        1000.0,
        1050.0,
        1100.0,
        1150.0,
        1200.0,
    ]
)

y_train = np.array([y + np.random.randint(-100, 100) for y in y_train])

print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train.shape = {x_train.shape}")
print(f"y_train.shape = {y_train.shape}")

if x_train.shape[0] != y_train.shape[0]:
    raise ValueError("x_train and y_train must have the same number of samples")

m = x_train.shape[0]
print(f"m = {m}")
```

```python
i = 4

x_i = x_train[i]
y_i = y_train[i]
print(f"x_{i} = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker="x", c="r")
plt.title("House Prices")
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price ($1000)")
plt.show()
```

```python
# f_wb(x) = w * x^(i) + b
w, b = 60, 100
print(f"w = {w}, b = {b}")
print(f"f(x) = {w} * x + {b}")
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b

    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)

plt.plot(x_train, tmp_f_wb, c="b", label="f(x)")
plt.scatter(x_train, y_train, marker="x", c="r", label="y")

x_i = 10.24
y_i = w * x_i + b
plt.scatter(x_i, y_i, marker="o", c="g", label="y_i", s=100, zorder=3)

plt.title("House Prices")
plt.xlabel("Size (1000 sqft)")
plt.ylabel("Price ($1000)")
plt.legend()
plt.show()
```
